```python
def concat_Table_78():
    
    path78='/Users/barbenshabat/Downloads/Datascience/project/Table_78/'
    name78='Table_78_Full-time_Law_Enforcement_Employees_by_State_by_City_201'
    file_name_table_78 = list_edited_name(name78)
    dfs_list = []
    
    for f in range(10):
        df78=pd.read_excel(path78+file_name_table_78[f], index_col=0) # read each file in the folder 
        if f==0: columns_name=df78.columns
        df78=city_value_column_cleaning(df78)
        df78=fit_columns_name(df78,columns_name)
    
        dfs_list.append(df78) # inserting each df to a list

    main_df = pd.concat(dfs_list,axis=0) # concanicating each df into main df 
    main_df.drop(['Unnamed: 0'], axis=1, inplace=True)
    main_df.to_excel('concat_table_78.xlsx',engine='xlsxwriter')
    return main_df
```

```python
# merging the 2 data sets table 8 and 78
# Law Enforcement & crime statistics

def merge_table8_and_table78():
    
    path_main_Table='/Users/barbenshabat/Downloads/Datascience/project/'
    name_Table_8='merge_table_8.xlsx'
    df8=pd.read_excel(path_main_Table+name_Table_8,index_col=0)
    df8=city_value_column_cleaning(df8)
    
    name_Table_78='concat_table_78.xlsx'
    df78=pd.read_excel(path_main_Table+name_Table_78,index_col=0)
    df78.drop(columns=['Total law enforcement employees','Total officers','Total civilians'], inplace=True)
    
    result = pd.merge(df8, df78, how="outer", on=['Year', 'State', 'City', 'Population'])
    #result.dropna(thresh=10, inplace=True)##think how to fill maybe by avg total police force
    result.to_excel('merge_T_8_and_78.xlsx',engine='xlsxwriter')
    return result
    
```

```python
# Editing the City Coulmn 

def city_value_column_cleaning(df):
    for city in df['City']:
        if city[len(city)-1].isnumeric(): 
            df.replace(city, city[:-1], inplace=True)
    return df
    
```

```python
# Edit file names 
def list_edited_name(name):
    name_list = []
    for i in range(10):
        name_list.append(name+str(i)+".xls_eddited.xlsx")
    return name_list
    
```

```python
# renameing the columns names
def fit_columns_name(df, columns_name):
    for i in range(0,len(columns_name)): df.rename(columns={df.columns[i]:columns_name[i]},inplace=True)
    return df
```

```python
## handling the nan values after the merge
def merge_table8_and_table78_nan_hendle(df):
    df.dropna(subset=['Total_Police force'], thresh=1, inplace=True)
    #df.fillna(0, inplace=True)
    df.reset_index(drop=True, inplace=True)
    df.to_excel('merge_T_8_and_78.xlsx',engine='xlsxwriter')
    return df
```
